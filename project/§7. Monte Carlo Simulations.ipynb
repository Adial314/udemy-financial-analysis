Monte Carlo simulations (MCS) attempt to model a system using a range of inputs from random variables (RVs). If a RV has a given standard deviation, the MCS will find best and worst case scenarios from within and without the standard deviation of each of its RVs. As the number of RVs increase, so too do the permutations which the MCS must consider.

The application of MCS results in a larger data set to study as new outcomes have been estimated using existing historical data. It is important to remember that MCS does not predict the future but simply provides an outcome range which encapsulates the most likely future outcomes.

Consider predicting the gross profit of a company using historical trends. We will synthesize data using normal distributions to represent the revenue streams of the company.


```python
# ---------------- IMPORT LIBRARIES ---------------- #

# Standard mathematics
import numpy as np

# Plotting and 3D axes tools
import matplotlib.pyplot as plt
%matplotlib inline

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')


# ---------------- SYNTHESIZING DATA ---------------- #

# Set seed
np.random.seed(0)

# Define Gaussian stats
revenue_mu  = 170   # Statistical mean
revenue_sigma = 20  # Statistical standard deviation

# Specify sample number
iterations = 1000

# Create distribution
revenue = np.random.normal(revenue_mu, revenue_sigma, iterations)


# ---------------- VISUALIZING DATA ---------------- #

# Plot distribution
plt.figure(figsize = (15, 6))
plt.plot(revenue)
plt.show()
```

```python
# Plot Gaussian
count, bins, ignored = plt.hist(revenue, 60, density=True)
plt.plot(bins, 1/(revenue_sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - revenue_mu)**2 / (2 * revenue_sigma**2) ), linewidth = 2, color = 'r')
plt.show()
```

We can confirm that the mean is 170 from both of our generated charts which indicates that our statistical models are accurately portraying our revenue.

We know that the cost of goods sold (COGS) can be represented by our revenue streams because COGS 60% of our revenue with a standard deviation of 10%.

```python
# ---------------- CREATING COGS ---------------- #

# Define COGS
COGS = - (revenue * np.random.normal(0.6, 0.1))

# Study COGS
print("COGS STDEV: ", COGS.std())
print("COGS MEAN:  ", COGS.mean())
```

```python
# Plot distribution
plt.figure(figsize = (15, 6))
plt.plot(COGS)
plt.show()
```
